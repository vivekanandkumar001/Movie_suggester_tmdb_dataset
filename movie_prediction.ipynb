```python
import kagglehub
import pandas as pd # Import the pandas library

# Download latest version
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")

print("Path to dataset files:", path)

# Use pandas to read the CSV file located at 'path'
movie = pd.read_csv(f"{path}/tmdb_5000_movies.csv")
credit=pd.read_csv(f"{path}/tmdb_5000_credits.csv")
# movie.sample(10) # Removed or commented out this line to prevent potential reassignment as a tuple
```

```python
# Merge the two dataframes
movie = movie.merge(credit, on='title')

# Now 'movie' is the merged dataframe, you can check its shape
print(movie.shape)

# And then display the head
movie.head()
```

```python
movie=movie[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
# Import the ast module to use literal_eval
import ast

def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L

movie['genres']=movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.head(1)
```

```python
def convert3(obj):

  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movie['cast'] = movie['cast'].apply(convert3)
```

```python
movie['cast']
```

```python
movie.head()
```

```python
movie['crew'][0]
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
  return L
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie
```

```python
movie['overview']=movie['overview'].apply(lambda x:x.split())
```

```python
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie
```

```python
movie['tags']=movie['overview']+movie['genres']+movie['keywords']+movie['cast']+movie['crew']
```

```python
movie['tags']
```

```python
new_df=movie[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1]) [1:6]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
  return
```

```python
recommend('Avatar')
```
